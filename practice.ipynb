```python
import numpy as np
import pandas as pd 
from matplotlib import pyplot as plt
from scipy import stats
```

```python
p = np.random.exponential(scale=1.0,size=10000) # population
```

```python
n = [2,4,8,10] # sample size
```

```python
f,a = plt.subplots(2,2,figsize=(12,8))
a = a.ravel()
for i,v in enumerate(n):
    m = [np.random.choice(p,size=n).mean() for _ in range(1000)]
    a[i].hist(m,bins=30,edgecolor="black")
    a[i].set_title(f"n={v}")
    a[i].set_xlabel("sample mean")
plt.show()
```
